```python
# 트리를 만들어 봅시다
def treeprinter(n):
    for i in range(n):
        if i < n - 5:
            print(" " * (n-i) + "*" * (2 * i +1))
        else: 
            print(" " * (n-1) + "*" * 3)

treeprinter(25)
```

```python
# 숫자 맞추기
import random

num_range = 15         # 1부터 입력한 수 사이의 임의의 수가 정답으로 설정됩니다. 수를 입력하세요
answer_num = random.randint(1,num_range)
guess_count = 0
max_chance =  3     # 정답을 맞출 수 있는 횟수를 정합니다.
while guess_count < max_chance:
    guess_num = int(input("답을 입력하세요: "))
    guess_count += 1
    if guess_num == answer_num:
        print("정답")
        break
    elif guess_count == max_chance and guess_num != answer_num:
        print(f"GAME OVER \n정답은 {answer_num}")
    elif guess_num < answer_num:
        print("업")
    elif guess_num > answer_num:
        print("다운")
```

```python
# 유-명한 로그인 문제
print("문제) 다음 아이디와 비밀번호를 참고하여 간단한 로그인 창을 만들어보세요.")
my_id_pw = {'professor': 'loveya486', 'hyempire': 'blood327'}
print(f"아이디, 비번 : {my_id_pw}\n")

while 1:
    print("아이디를 입력하세요 :")
    my_id = input()                                 #아이디를 입력 받습니다.
    if my_id in my_id_pw:                           #아이디가 dictionary에 있는지 확인합니다.
        print("비밀번호를 입력하세요 : ")
        my_pw = input()                             #아이디가 있으면 비밀번호를 입력 받습니다.
        if my_pw == my_id_pw[my_id]:                #비밀번호가 해당 아이디의 value 값과 같은지 확인합니다.
            print("환영합니다!^^")
            break                                   #같으면 환영해준 후 while문을 빠져 나갑니다.
        else:
            print("비밀번호를 다시 확인해주세요.")       #다르면 다시 처음으로 돌아갑니다.
            continue
    else:
        print("아이디를 다시 확인해주세요.")
        continue
```

```python
# 유-명한 구구단 문제
print("문제) 구구단 2단에서부터 9단까지 출력하세요.\n")

for i in range(2, 10):
    for v in range(1, 10):
        print(f"{i} x {v} = {i * v}", end = "   ")
    print()
```

```python
# 작년 기출 - 30(임의의 수)보다 작거나 같은 짝수와 홀수 곱 구하고 대소 비교

even_mul = 1
odd_mul = 1
for i in range(1, 31):  
    if i % 2 == 0:
        even_mul *= i
    else:
        odd_mul *= i
if even_mul > odd_mul:
    result = f"짝수곱 {even_mul}이 홀수곱 {odd_mul}보다 크다"
elif odd_mul > even_mul:
    result = f"홀수곱 {odd_mul}이 짝수곱 {even_mul}보다 크다"
else:
    result = f"짝수곱{even_mul}과 홀수곱 {odd_mul}의 크기는 같다"
print(result)
```

```python
# 리스트 안의 수들을 sort없이 오름차순으로 배열

a_list = [8, 5, 7, 3, 11, 15, 19, 205, 34, 25, 88]
for i in range(len(a_list)-1):
    for j in range(i+1, len(a_list)):
        if a_list[i] > a_list[j]:
            a_list[i], a_list[j] = a_list[j], a_list[i]
        
print(a_list)
```
