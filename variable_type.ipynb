1. 자료형

  자료형에 대한 예제들을 모아놓는 공간입니다. 
  각 자료형별로 모아주시되 여러 자료형이 겹치는 경우는 어떻게 분류할지 논의해 봅시다. 아예 맨 뒤로 따로 빼도 좋을 것 같고요.

  * 각 예제는 좌측 상단의 +코드 버튼을 클릭하여 예제별로 각각의 코드로 구성되게 하는 방식을 제안합니다.

1-1. number

```python
# 예시
a = 2
b = 10
c = a ** b
print(c)
```

```python
# 요렇게 문제별로 각각..
a = 2
b = 10
c = a / b
print(c)
print(type(c))
```

```python
# 나름 boolean 문제,,
print("문제) 자연수 20이 홀수인지 구별할 수 있는 방법을 구현해보세요.\n")

num = 20                    # 자연수인 정수를 선언했습니다.
judge = bool(num % 2)       # 2로 나눴을 때의 나머지가 0이면 짝수이므로, 홀수면 True, 짝수면 False가 출력되게 하였습니다.

print(f"{num}은 홀수가 맞습니까?")
print(f"{judge}입니다.")
```

1-2. string

```python
print("문제) 세상에서 가장 귀여운 사람의 이름을 100번 출력해보세요. \n")

my_name = "혜미"

print(f"세상에서 가장 귀여운 사람의 이름은 {my_name}입니다.")
print(my_name * 100)        #string은 곱하기를 하면 같은 문자열이 반복되며 출력됩니다.
```

1-3. tuple

```python
print("문제) (1, 5, 5, 2, 3, 2, 4, 6)과 (5, 1, 5, 8, 9, 3, 9)의 교집합을 구하시오.")
print("\t   (단, 중복된 숫자는 한 번만 쓰시오)\n")\

a = {1, 5, 5, 2, 3, 2, 4, 6}        #중복된 숫자는 없애기 위해 Set으로 선언했습니다.
b = {5, 1, 5, 8, 9, 3, 9}
c = a & b

print(f"위 숫자들의 교집합은 {c}입니다.")
```

1-4. list

```python
print("문제) 혜미의 시험성적을 list로 선언하고 인덱스를 이용하여 평균 점수를 구해보세요.\n")

score = [85, 87.5, 95, 100]                                             # 4개의 시험 성적을 list로 선언했습니다.
avg = (score[0] + score[1] + score [2] + score[3]) / len(score)         # 성적을 모두 더한 뒤 리스트 요소의 개수로 나눴습니다.

print(f"혜미의 시험성적은 {score}입니다.")
print(f"\t국어 : {score[0]}점 / 수학 : {score[1]}점 / 영어 : {score[2]}점 / 미술 : {score[3]}점")
print(f"시험성적 평균은 {avg}점 입니다.")
```

```python
# 이거 90년대생도 같이 구하려면 홀짝으로 구해야 해요
print("문제) 주민번호(000327-4XXXXXX)에서 성별을 구별해보세요.")

id_num = '000327-4XXXXXX'       # 주민번호를 문자열로 선언했습니다.
gender = int(id_num[7])         # 숫자의 홀짝을 구별하기 위해 int로 형변환 했습니다.
judge = bool(gender % 2)        # 짝수면 여자(False), 홀수면 남자(True)입니다.
print(f"주민번호가 {id_num}인 사람은 남자입니까?")
print(f"{judge}입니다.")
```

1-5. dictionary

```python
print('''문제) '나이와 성별을 dictionary로 선언하고, 거기에 전화번호를 추가해보세요.\n''')

personal_dict = {'age' : 22, 'gender' : 'female'}

print(f"나이와 성별 정보입니다 : {personal_dict}")

personal_dict['phone_num'] = '010-1234-5678'        #dict는 인덱스로 기존에 없던 Key를 지정하여 값을 부여하면 Key가 추가됩니다.

print(f"전화번호 정보를 추가한 결과입니다 : {personal_dict}")
```
