함수의 정의 방법
함수는 def 키워드로 정의하며, def 함수이름(): 으로 정의 할 수 있습니다.
함수 안에는 어떤 코드라도 있어야 합니다. 만약 아무것도 하지 않는 함수가 필요하다면, pass를 사용하면 됩니다.
반환이 필요하다면 return을 이용합니다.

```python
def func():
  pass

func()
```

```python
def func_2():
  return 1, 2, 3

def func_3():
  return (1, 2, 3)

tmp1 = func_2()
tmp2 = func_3()

print(f"{tmp1} : Type - {type(tmp1)}")
print(f"{tmp2} : Type - {type(tmp2)}")
```

Mutable(가변형) Vs Immutable (불변형)

Mutable Type: Numeric Type(int, float complex), Immutable Sequences(string tuple frozenset bytes)

Immutable Type: Mutable Sequences(list  bytearray), Set Type(set), Mapping Type(dict), Class, Class Instance


```python
def func4(mutable, immutable, mutable_id, immutable_id):
  mutable.append(3)
  immutable += "string"
  print(f"가변형 {id(mutable) == mutable_id }")
  print(f"불변형 {id(immutable) == immutable_id}")

string_immutable = "string"
list_mutable = [1, 2]

func4(list_mutable, string_immutable, id(list_mutable), id(string_immutable))
print(f"List: {list_mutable}")
print(f"String: {string_immutable}")
```

함수에 인자를 받을 때 메모리와 성능을 위해 크기가 큰 값들은 값을 복사하지 않고 참조합니다. 참조하는 값들은 함수 안에서 값이 바뀌면 함수 밖에서도 값이 바뀝니다.

함수에서 인자가 받지 못하면, 기본값을 사용하게 할 수 있습니다.

```python
def func5(a, default_value=5):
  print(f"{default_value} - 기본값은 5입니다.")

func5(1)
func5(1, 2)
```

print() 함수를 보면 인자의 갯수가 정해지지 않아도 인자를 여러개 받을 수 있습니다.
*매개변수명 으로 사용합니다.
인자들이 튜플로 묶여서 사용가능합니다.

```python
def func6(*args):
  print(args)
func6("1", [2], 3)
```

여러개의 키워드 인자값들을 받을 수 있습니다.

```python
# 더 큰 수 반환하는 함수

def get_max(a, b):
    if a > b:
        my_result = f"{a}가 {b}보다 크다"
    elif a < b:
        my_result = f"{b}가 {a}보다 크다"
    else:
        my_result = f"{a}와 {b}의 값은 같다"
    return my_result
```

```python
# 유-명한 구구단 함수 문제
print(f"문제) 구구단을 출력해주는 함수를 만들고 정수를 입력받아 구구단을 출력하세요")
num = int(input("구구단을 수행할 숫자를 입력해주세요 : "))
print("")

def gugudan(num):
    for i in range(1, 10):
        print(f"{num} x {i} = {num * i}")

gugudan(num)
```

```python
# 단위 바꾸는 함수 문제
print(f"문제) default를 이용하여 미터를 센티미터로 바꾸는 함수를 만드세요")

def meter_to_centimeter(meter_num, measurement = 100):          #default 값으로 100을 넣어주었습니다
    centi_num = meter_num * measurement
    return centi_num

meter_num =float(input("센티미터로 단위를 변환할 미터값을 입력하세요 : "))
print("")
print(f"{meter_num}m = {int(meter_to_centimeter(meter_num))}cm")
```

```python
print(f"문제) 문자열을 입력 받고, 문자 사이사이에 ', '를 더해주는 함수를 만들세요\n")

def add_comma(my_string):
    str_comma = ""
    for i in my_string:
        str_comma += i + ", "
    return str_comma[:-2]

my_string = input("문자열을 입력하세요 : ")
print(add_comma(my_string))
```
